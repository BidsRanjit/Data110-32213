# Creating and Customizing Bar Charts with Matplotlib

```python

from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
plt.style.use("ggplot")  # Sets the style for the plots
    
```

### Basic Bar Chart

```python

x = [1, 2, 3, 4]
y = [10, 8, 12, 16]
plt.bar(x, y, label='Week 1')
plt.legend()
plt.show()
    
```

### Comparing Two Sets of Data

```python

x = ['p1', 'p2', 'p3', 'p4']
y = [10, 8, 12, 16]
z = [12, 7, 8, 23]
plt.bar(x, y, label='Week 1')
plt.bar(x, z, label='Week 2', bottom=y)  # Stacks z on top of y
plt.legend()
plt.show()
    
```

### Adjusting Bar Positions

```python

x_indexes = np.arange(len(x))
plt.bar(x_indexes - 0.125, y, width=0.25, label='Week 1')
plt.bar(x_indexes + 0.125, z, width=0.25, label='Week 2')
plt.xticks(ticks=x_indexes, labels=x)
plt.legend()
plt.show()
    
```

### Visualizing Real Data

```python

# Load the dataset
df = pd.read_csv('/Users/Rebin/Downloads/happiness (1).csv')
df.head()  # Shows the first few rows
df.info()  # Provides information about the dataframe

# Basic visualization
plt.figure(figsize=(25, 25))
plt.barh(df['country'], df['Happiness Score'])
plt.show()
    
```

### Enhancing the Visualization

```python

df_sorted = df.sort_values(by='Happiness Score', ascending=True)
plt.style.use('fivethirtyeight')
plt.figure(figsize=(10, 20))
plt.barh(df_sorted['country'], df_sorted['Happiness Score'])
plt.xlabel('Happiness Score')
plt.ylabel('Country')
plt.title('Happiness Score by Country')
ax = plt.gca()
ax.tick_params(axis='y', labelsize=8)
plt.show()
    
```

### Coloring Bars Based on Value

```python

df['Score Quartile'] = pd.qcut(df['Happiness Score'], 4, labels=['Q1', 'Q2', 'Q3', 'Q4'])
color_mapping = {'Q1': 'red', 'Q2': 'orange', 'Q3': 'lightblue', 'Q4': 'green'}
df['Color'] = df['Score Quartile'].map(color_mapping)
plt.figure(figsize=(10, 20))
plt.barh(df['country'], df['Happiness Score'], color=df['Color'])
plt.xlabel('Happiness Score')
plt.ylabel('Country')
plt.title('Happiness Score by Country')
ax = plt.gca()
ax.tick_params(axis='y', labelsize=8)
plt.show()
    
```
